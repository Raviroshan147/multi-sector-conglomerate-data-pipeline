```python
# ==============================================================================
# PRODUCTION-GRADE DATA ENGINEERING ENGINE FOR 7 SECTORS (STAR SCHEMA)
# AUTHOR: RAVI ROSHAN KUMAR (DATA ANALYST)
# ==============================================================================

try:
    import openpyxl
    from faker import Faker
except ImportError:
    !pip install faker openpyxl pandas

import datetime
import os
import random
import shutil
import pandas as pd
from faker import Faker

# Setup Reproducible Data Engine
fake = Faker(['en_IN'])
Faker.seed(2026)
random.seed(2026)

base_dir = "Star_Schema_Output"
os.makedirs(base_dir, exist_ok=True)
print(f"⏳ Generating 14 Fact & Dim Tables for 7 Sectors inside '{base_dir}'...\n")

# 1. DIGITAL MARKETING
print("⚡ Processing: Digital Marketing Agency...")
client_keys = [f"CLT-{i:03d}" for i in range(1, 21)]
pd.DataFrame({
    "Client_Key": client_keys,
    "Client_Name": [fake.company() for _ in client_keys],
    "Industry": [random.choice(["E-Commerce", "EdTech", "FinTech", "Automobile", "Healthcare"]) for _ in client_keys],
    "Account_Manager": [fake.name() for _ in client_keys]
}).to_excel(f"{base_dir}/Dim_Marketing_Clients.xlsx", index=False)

m_channels = ["Google Search Ads", "Meta Instagram Ads", "LinkedIn B2B Ads", "SEO Optimization", "YouTube Video Ads"]
m_strat = ["Brand Awareness", "Lead Generation", "Retargeting", "Festive Offer"]
m_fact = []
for i in range(365):
    dt = (datetime.date(2025, 6, 1) + datetime.timedelta(days=i)).strftime("%Y-%m-%d")
    for c_key in random.sample(client_keys, random.randint(4, 8)):
        for ch in m_channels:
            if random.random() > 0.4:
                sp = round(random.uniform(1000, 25000), 2)
                imp = random.randint(10000, 400000)
                cl = int(imp * random.uniform(0.02, 0.07))
                conv = int(cl * random.uniform(0.01, 0.04))
                rev = round(conv * random.uniform(80, 200), 2)
                m_fact.append([f"CAM-{random.randint(10000, 99999)}", dt, c_key, ch, random.choice(m_strat), imp, cl, conv, sp, rev])
pd.DataFrame(m_fact, columns=["Campaign_ID", "Date", "Client_Key", "Marketing_Channel", "Campaign_Strategy", "Impressions", "Clicks", "Conversions", "Ad_Spend_USD", "Revenue_USD"]).to_excel(f"{base_dir}/Fact_Marketing_Performance.xlsx", index=False)

# 2. HOSPITAL MANAGEMENT
print("⚡ Processing: Hospital Management...")
depts = ["Cardiology", "Neurology", "Pediatrics", "Orthopedics", "General Medicine"]
doc_keys = [f"DOC-{i:03d}" for i in range(1, 11)]
pd.DataFrame({
    "Doctor_Key": doc_keys,
    "Doctor_Name": [f"Dr. {fake.name()}" for _ in doc_keys],
    "Department": [random.choice(depts) for _ in doc_keys],
    "Consultation_Fee": [random.choice([600, 1000, 1500]) for _ in doc_keys]
}).to_excel(f"{base_dir}/Dim_Hospital_Doctors.xlsx", index=False)

ailments = ["Acute Hypertension", "Chronic Migraine", "Severe Viral Fever", "Ligament Tear", "Type-2 Diabetes", "Gastric Ulcer"]
h_fact = []
for i in range(365):
    dt = (datetime.date(2025, 6, 1) + datetime.timedelta(days=i)).strftime("%Y-%m-%d")
    for _ in range(random.randint(8, 15)):
        dk = random.choice(doc_keys)
        fee = 1000 if dk == "DOC-001" else 600
        lab = round(random.uniform(500, 5000), 2)
        phr = round(random.uniform(200, 3500), 2)
        h_fact.append([f"PAT-{random.randint(10000, 99999)}", dt, fake.name(), random.randint(1, 80), random.choice(["Male", "Female"]), dk, random.choice(ailments), fee, lab, phr, round(fee+lab+phr, 2)])
pd.DataFrame(h_fact, columns=["Patient_ID", "Admission_Date", "Patient_Name", "Age", "Gender", "Doctor_Key", "Diagnosis", "Consultation_Fee", "Lab_Charges", "Pharmacy_Bill", "Total_Bill_INR"]).to_excel(f"{base_dir}/Fact_Hospital_Admissions.xlsx", index=False)

# 3. SCHOOL PERFORMANCE
print("⚡ Processing: School Management...")
stu_keys = [f"STU-{i:04d}" for i in range(1001, 1121)]
pd.DataFrame({
    "Student_Key": stu_keys,
    "Student_Name": [fake.name() for _ in stu_keys],
    "Grade": [random.choice([f"Grade {g}" for g in range(6, 11)]) for _ in stu_keys],
    "Section": [random.choice(["Section A", "Section B"]) for _ in stu_keys],
    "Annual_Fees": [random.choice([50000, 65000, 80000]) for _ in stu_keys]
}).to_excel(f"{base_dir}/Dim_School_Students.xlsx", index=False)

subs = ["Mathematics", "Science", "English", "Social Science", "Computer Science"]
s_fact = []
for sk in stu_keys:
    att = round(random.uniform(70, 99), 1)
    for sb in subs:
        s_fact.append([sk, sb, random.randint(40, 98), att, random.choice(["Fully Paid", "Partially Paid", "Pending"])])
pd.DataFrame(s_fact, columns=["Student_Key", "Subject", "Exam_Score_Pct", "Attendance_Rate", "Fee_Status"]).to_excel(f"{base_dir}/Fact_School_Academics.xlsx", index=False)

# 4. CA FIRM
print("⚡ Processing: CA Firm Management...")
corp_keys = [f"CORP-{i:03d}" for i in range(1, 25)]
pd.DataFrame({
    "Corporate_Key": corp_keys,
    "Company_Name": [fake.company() for _ in corp_keys],
    "Business_Structure": [random.choice(["Private Ltd", "LLP", "Partnership", "Proprietorship"]) for _ in corp_keys]
}).to_excel(f"{base_dir}/Dim_CA_Clients.xlsx", index=False)

srv_costs = {"GST Filing": 4000, "ITR-6 Return": 8500, "Corporate Audit": 55000, "TDS Compliance": 5000}
ca_fact = []
for i in range(365):
    dt = (datetime.date(2025, 6, 1) + datetime.timedelta(days=i)).strftime("%Y-%m-%d")
    if random.random() > 0.4:
        for _ in range(random.randint(1, 3)):
            srv = random.choice(list(srv_costs.keys()))
            base = srv_costs[srv]
            ca_fact.append([f"INV-{random.randint(50000, 99999)}", dt, random.choice(corp_keys), srv, base, round(base*0.18, 2), random.choice(["Paid", "Unpaid"]), f"CA {fake.name()}"])
pd.DataFrame(ca_fact, columns=["Invoice_Number", "Billing_Date", "Corporate_Key", "Service_Type", "Professional_Fees", "GST_18_Pct", "Payment_Status", "Assigned_Partner"]).to_excel(f"{base_dir}/Fact_CA_Invoices.xlsx", index=False)

# 5. COAL TRADERS (DHANBAD)
print("⚡ Processing: Coal Traders (Dhanbad Market)...")
c_grades = ["G4 Grade (High CV)", "G7 Grade (Semi-Coking)", "G11 Grade (Power Sector)", "G13 Grade (Slurry)"]
c_prices = {"G4 Grade (High CV)": 6500, "G7 Grade (Semi-Coking)": 4900, "G11 Grade (Power Sector)": 3600, "G13 Grade (Slurry)": 2400}
buyers = ["ACC Cement Plant", "Maithon Power", "Damodar Steel", "Rungta Mines Ltd"]
ct_fact = []
for i in range(365):
    dt = (datetime.date(2025, 6, 1) + datetime.timedelta(days=i)).strftime("%Y-%m-%d")
    if random.random() > 0.3:
        for _ in range(random.randint(1, 3)):
            grd = random.choice(c_grades)
            qty = round(random.uniform(50, 800), 2)
            prate = round(c_prices[grd] * random.uniform(0.96, 1.04), 2)
            srate = round(prate * random.uniform(1.08, 1.15), 2)
            cess = round(qty * 400, 2)
            ct_fact.append([f"DL-{random.randint(1000, 9999)}", dt, grd, qty, f"{fake.city()} Siding", random.choice(buyers), prate, srate, round(prate*qty, 2), round(srate*qty, 2), cess])
pd.DataFrame(ct_fact, columns=["Deal_ID", "Transaction_Date", "Coal_Grade", "Quantity_Tons", "Source_Siding", "Buyer_Company", "Purchase_Rate", "Selling_Rate", "Gross_Purchase_Cost", "Gross_Revenue", "Clean_Energy_Cess"]).to_excel(f"{base_dir}/Fact_Coal_Trading_Deals.xlsx", index=False)

# 6. COAL TRANSPORTER
print("⚡ Processing: Coal Transporter Logistics...")
v_types = ["10-Wheeler Tipper", "12-Wheeler Ashok Leyland", "16-Wheeler Hyva"]
routes = ["Katras Siding to Maithon Power", "Jharia Colliery to Bokaro Steel", "Bastacola Dump to Siding"]
t_fact = []
for i in range(365):
    dt = (datetime.date(2025, 6, 1) + datetime.timedelta(days=i)).strftime("%Y-%m-%d")
    for _ in range(random.randint(4, 10)):
        vt = random.choice(v_types)
        wt = round(22 * random.uniform(0.95, 1.05), 2)
        gross = round(wt * 450, 2)
        dsl = round(gross * 0.42, 2)
        t_fact.append([f"TRIP-{random.randint(100000, 199999)}", dt, f"JH-10-{fake.lexify('??').upper()}-{random.randint(1000, 9999)}", vt, random.choice(routes), wt, gross, dsl, random.randint(1000, 2000)])
pd.DataFrame(t_fact, columns=["Trip_ID", "Trip_Date", "Truck_Number", "Vehicle_Type", "Route_Details", "Net_Weight_Tons", "Gross_Freight", "Diesel_Expense", "Toll_Loading_Charges"]).to_excel(f"{base_dir}/Fact_Coal_Logistics_Trips.xlsx", index=False)

# 7. CHEMIST & DRUG SHOP (FLAGSHIP MASTER)
print("⚡ Processing: Chemist & Drug Shop...")
med_keys = [f"MED-{i:03d}" for i in range(1, 9)]
med_names = ["Dolo 650mg", "Brufen 400mg", "Amoxil 500mg", "Azithral 500mg", "Glycomet 500mg", "Lipitor 10mg", "Pan-40", "Cetzine 10mg"]
med_cats = ["Analgesic", "NSAID", "Antibiotic", "Antibiotic", "Anti-Diabetic", "Cardiovascular", "Gastrointestinal", "Antihistamine"]
pd.DataFrame({
    "Product_Key": med_keys,
    "Medicine_Name": med_names,
    "Category": med_cats,
    "Cost_Price_Per_Unit": [12.50, 14.00, 70.00, 80.00, 22.00, 55.00, 38.00, 7.50],
    "Selling_Price_Per_Unit": [16.00, 21.50, 95.00, 120.00, 38.00, 85.00, 60.00, 12.00]
}).to_excel(f"{base_dir}/Dim_Pharmacy_Products.xlsx", index=False)

p_fact = []
order_id = 550001
for i in range(365):
    dt = (datetime.date(2025, 6, 1) + datetime.timedelta(days=i)).strftime("%Y-%m-%d")
    for _ in range(random.randint(25, 45)):
        pk = random.choice(med_keys)
        qty = random.randint(1, 5)
        batch = f"BCH-{fake.lexify('???').upper()}-{random.randint(100, 999)}"
        exp = (datetime.date(2026, 6, 1) + datetime.timedelta(days=random.randint(100, 500))).strftime("%Y-%m-%d")
        p_fact.append([f"TXN-{order_id}", dt, random.choice(["Walk-in", "Regular"]), pk, batch, exp, qty])
        order_id += 1
pd.DataFrame(p_fact, columns=["Transaction_ID", "Date", "Customer_Type", "Product_Key", "Batch_No", "Expiry_Date", "Quantity_Sold"]).to_excel(f"{base_dir}/Fact_Pharmacy_Sales.xlsx", index=False)

# Zipping the files
shutil.make_archive(base_dir, 'zip', base_dir)
print("\n" + "="*60)
print("🎉 SUCCESS! ALL 14 TABLES GENERATED IN STAR SCHEMA ARCHITECTURE!")
print(f"👉 '{base_dir}.zip' फाइल डाउनलोड के लिए तैयार है।")
print("="*60)
```
